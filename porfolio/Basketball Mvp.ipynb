## Feature Engineering 

dari data diatas terdapat fitur FG% yang biasanya dipakai untuk mengukur effisiensi player
namun fitur tersebut tidak mempertimbangkan jumlah FT yang pemain tembak, FT telah di setujui oleh coach and pemain basket
merupakan hal yang paling efisien dalam sisi offence basket. Jika hanya melihat FG% untuk menentukan efisiensi player maka
data tersebut kurang maksimun. Maka dari itu kita membuat sebuah metrik yang mempertimbangkan FT

#Variabel yang diperlukan:
- FGA
- FTA
- rata rata posesi FT dalam 1 game (FT shooting, 3s violation, Technical Freethrow)
 (PTS * 0.5) / (FGA + (0.45 * FTA))



TE% dapat memberikan gambaran yang lebih lengkap tentang efisiensi, Setelah membuat fitur TE, ini dapat dimasukkan ke dalam model machine learning untuk menganalisis kinerja pemain 
dan dapat digunakan sebagai fitur yang berguna untuk memprediksi efisiensi dan kontribusi seorang pemain dalam tim basket.


```python
data['TE%'] = data['PTS']*0.5/(data['FGA']+ (0.45*data['FTA'])) #0.45 dari rata rata possession free throw
```

```python
#membuat fitur rata rata metrik liga 
```

```python
def league_average(data, cols, new_cols):
        data[new_cols] = data[cols].mean()
        return data
```

```python
grouped_data = data.groupby('Year', group_keys= True)
```

```python
cols = ['G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'TE%', 'W']
new_cols = [f"League Average {i}" for i in cols]
```

```python
data_w_averages = data.groupby('Year', group_keys= True).apply(lambda x : league_average(x, cols, new_cols) )
```

```python
from sklearn import svm
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = modern_era[efficiency + bruh + intangibles]
y = modern_era['Share']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
regr = svm.SVR()
```

```python
regr.fit(X_train,y_train)
```

```python
predictions = regr.predict(X_test)
```

```python
predictions
```

```python
from sklearn import metrics
```

```python
mae_value = metrics.mean_absolute_error(y_test, predictions)
mse_value =  metrics.mean_squared_error(y_test, predictions)
rmse_value =np.sqrt(metrics.mean_squared_error(y_test, predictions))
```

```python
metrics_names = ['MAE', 'MSE', 'RMSE']
metrics_values = [mae_value, mse_value, rmse_value]

plt.figure(figsize=(8, 6))
plt.bar(metrics_names, metrics_values, color=['blue', 'green', 'red'])
plt.xlabel('Metrics')
plt.ylabel('Value')
plt.title('Performance Metrics')
plt.show()
```

```python
print('MAE: ',mae_value)
print('MSE: ',mse_value)
print("RMSE: ",rmse_value)
```

```python
data = np.arange(len(y_test))
plt.figure(figsize=(10, 6))
plt.plot(data, y_test, label='Value Asli', marker='o', linestyle='-')
plt.plot(data, predictions, label='Value Prediksi', marker='x', linestyle='--')
plt.title('Prediksi vs Value Asli')
plt.xlabel('Data Points')
plt.ylabel('Values')
plt.xticks(data)
plt.legend()
plt.grid(True)
plt.show()
```
