```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
import re
```

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2023.html'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
game_log = soup.find_all('table')[4]
```

```python
game_log.find_all('th')
```

```python
cols = [kolom.text.strip() for kolom in game_log]
```

```python
cols[1:5]
```

```python
cols[5]
```

```python
cols = cols[5].split(" ")
```

```python
cols
```

```python
df = pd.DataFrame(columns= cols)
```

```python
df
```

```python
rows = game_log.find_all('tr')
```

```python
rows
```

```python
i = 1
for row in rows[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text for data in row_data]
    individual_row_data
    length = len(df)
    df.loc[length] = [i] + individual_row_data
    i = i +1
    
```

```python
df
```

```python
df.to_csv("2023 Season.csv")
```
